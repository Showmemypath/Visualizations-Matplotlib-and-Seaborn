```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
plt.style.use('dark_background')
```

```python
x = np.linspace(0,2*np.pi,100)
sin_x = np.sin(x)
y = np.linspace(0,4*np.pi,100)
cos_x = np.cos(y)

fig = plt.figure(figsize=(10,4))

plt.ylim([-2,2])
plt.xlim(-5,14)
# or plt.axis([xlim,ylim]) 

plt.plot(x,sin_x ,'|',label="Sin",color="grey",linestyle='-') # It used both linestyle='-' and '|' marker too
plt.plot(y,cos_x,'+',label="Cos",color="white",linewidth=10) # linewidth doesn't work because '+' is a marker style 
# marker style plots points at each data point , so there is no line to apply line width to

plt.fill_between(x,sin_x,cos_x,color='yellow',alpha=0.3)

plt.legend(loc="lower right")
fig.savefig('first fig save')
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)

fig = plt.figure(figsize=(10,4))

fig2 = plt.axes()
fig2.plot(x,np.log(x))

#Can set anything about the figure inside .set
fig2.set(ylim=(-2,2) , xlabel='X-axis', ylabel='Y-axis', title='Basic graph')

plt.legend(loc="lower right")
plt.show()
```

```python
rng = np.random.RandomState(0)
```

```python
# multiple scatter plots

x = rng.random(100)
y = rng.random(100)
plt.grid(True)
plt.scatter(x,y, c = 'white',alpha=.5,s=30)
plt.scatter(x+1,y+1,alpha=.5,s=30)
plt.show()
```

```python
plt.style.use('Solarize_Light2')
```

```python

plt.figure(figsize=(10,5))
colors = rng.random(100)
sizes = rng.random(100) * 1000

plt.scatter(x,y, c = colors,alpha=.8,s=sizes,cmap='viridis',edgecolors='black',label='Random numbers from a uniform distribution')

plt.xlabel='X-axis'
plt.ylabel='Y-axis'
legend = plt.legend()
for text in legend.get_texts():
    text.set_color('black')

plt.colorbar()
plt.title('Scatter plot with different size and opacity')
plt.show()
```

```python
plt.figure(figsize=(8,5))
x = np.arange(0,11)
y = np.arange(1,2*x[-1] + 2,2)
plt.plot(x,y,'o-',c='orange')
#plt.annotate('(1,3)',(1,3),textcoords='offset points',xytext=(0,10),ha='center')
for xi, yi in zip(x, y):
    plt.annotate(f'({xi}, {yi})',               #text to put
                 (xi, yi),                      #point to put the text at
                 textcoords="offset points",    #treat the point target as origin so that we can position the text using xytext coordinates
                 xytext=(0, 10),
                 ha='center')                   #horizontally center text whereever the text is places
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('Is this a line graph?')
plt.show()
```

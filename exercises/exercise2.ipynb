```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\fbase\\All_vscode_projects\\mlproject\\src\\nyc_311_sample.csv")
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-muted')
```

```python
data = df.complaint_type.value_counts()[:10]
plt.figure(figsize=(10,6))
plt.bar(data.index,data)
plt.xlabel('Complaint types')
plt.ylabel('Count')
plt.title('Top10 Complaint types')
plt.xticks(rotation=45)
plt.show()
```

```python
data = df.agency_name.value_counts()
plt.barh(data.index,data)
plt.xlabel('agency_name')
plt.ylabel('Count')
plt.title('Which agency handled the most requests')
plt.show()
```

```python
cmap = cm.get_cmap('cool',6)
```

```python
data = df.borough.value_counts()
plt.pie(labels=data.index,x=data,colors=[cmap(i) for i in range(1,7)])
plt.title('Service requests proportions per borough')
plt.show()
```

```python
df[['created_date','closed_date']] = df[['created_date','closed_date']].astype('datetime64[ns]')
df['response_time'] = df['closed_date'] - df['created_date']
```

```python
df['created_time'] = df.created_date.dt.time
df['closed_time'] = df.closed_date.dt.time
```

```python
data = df.created_date.dt.day.value_counts()
data = data[np.sort(data.index)]
```

```python
plt.figure(figsize=(8,6))
plt.grid(True)
plt.plot(data)
plt.xlabel('date')
plt.ylabel('Count')
plt.title('Service request amount trend over days')
plt.xticks(np.arange(12,32))
plt.show()
```

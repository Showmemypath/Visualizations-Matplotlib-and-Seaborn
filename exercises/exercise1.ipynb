```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1,11)
y = x*2

plt.plot(x,y,'*',color='blue',linestyle='dotted')
plt.title('Squares')
plt.xlabel('x-axis')
plt.ylabel('y-axes')

plt.show()
```

```python
y1 = x^2
y2 = x^3

plt.grid(True)
plt.plot(x,y1,label='squared')
plt.plot(x,y2,label='cubed')
plt.legend()
plt.show()
```

```python
x = np.random.randint(0,100,size=(50,))
y = np.random.randint(0,100,size=(50,))
plt.grid(True)
plt.scatter(x,y,marker='*',c='black',alpha=0.9,s=5)
plt.show()
```

```python
plt.bar(['Math','Physics','Chemistry','Biology','English'],
        list(range(50,100,10))
        ,color='pink',
        width=0.5,
        bottom=0,
        #,align='center' # center is default
        )
plt.xlabel('x-axis')
plt.title('First bar plot')
plt.xticks([0,1,2,3,4],['meth','Phy','Chem','Bio','Eng']) # The first list is for giving position
plt.show()
```

```python
import matplotlib.cm as cm
cmap = cm.get_cmap('viridis',4)

# Create a DataFrame like this:
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Q1': [10, 15, 20, 25],
    'Q2': [12, 18, 22, 28],
    'Q3': [11, 16, 21, 27],
    'Q4': [13, 19, 23, 29]
}
df = pd.DataFrame(data).set_index('Product')


# Question: 
# Create a stacked bar plot where:
# - Each bar represents a product
# - Q1-Q4 are stacked on top of each other
# - Use different colors for each quarter
# - Add percentage labels on each segment #try plt.text
# - Include a title and axis labels

plt.bar(df.index,df.Q1,bottom=0,color=cmap(0))
plt.bar(df.index,df.Q2,bottom=df.Q1,color=cmap(1))
plt.bar(df.index,df.Q3,bottom=df.Q1+df.Q2,color=cmap(2))
plt.bar(df.index,df.Q4,bottom=df.Q1+df.Q2+df.Q3,color=cmap(3))
plt.text(1,50,s="hi",color='black')

plt.xlabel('categories')
plt.ylabel('count')
plt.title('Boxplot?')
plt.show()

```

Create a bar chart for:

Students = A, B, C, D
Marks = 65, 72, 58, 80

Then try:
• different colors
• width of bars
• add a title

```python
import matplotlib.cm as cm
```

```python
cmap = cm.get_cmap('cool',4)
Students = ['A','B','C','D']
Marks = [65,72,58,80]

plt.bar(Students,Marks,width=0.3,color=[cmap(i) for i in range(4)])
plt.title('Student Marks')

plt.show()
```

```python
cmap = cm.get_cmap('twilight_shifted')
Students = ['A','B','C','D']
Marks = [65,72,58,80]

plt.barh(Students,Marks,height=0.5,color=[cmap(i) for i in range(4)])
plt.title('Student Marks')

plt.show()
```

```python
cmap = cm.get_cmap('cool',4)
what = ['Sleep','Study','Leisure','Other']
hours = [8,6,4,6]

plt.pie(hours,labels=what,autopct='6',explode=[0.5,0.5,0.5,0.5],colors=[cmap(i) for i in range(4)])

plt.show()
```
